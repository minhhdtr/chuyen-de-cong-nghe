```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta

#lay ngày hiện tại 
today = date.today()

d1 = today.strftime("%Y-%m-%d")
#ngày cuối cùng lấy dữ liệu
end_date = d1
d2 = date.today() - timedelta(days=1826) # lấy ngày từ 5 năm trước đây 
d2 = d2.strftime("%Y-%m-%d")
#ngày bắt đầu dữ liệu
start_date = d2


#lấy dữ liệu bitcoin từ yahoo finance 
data = yf.download('BTC-USD', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
# phân loại dữ liệu 
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data_last_30days = data[-30:]
data_last_year = data[-365:]


# Date: Ngày giao dịch hoặc ngày mà dữ liệu được ghi lại. Thường được sử dụng làm chỉ mục chính trong phân tích chuỗi thời gian.
# Open: Giá mở cửa của tài sản trong phiên giao dịch, tức là giá đầu tiên mà tài sản được giao dịch trong ngày.
# High: Giá cao nhất mà tài sản đạt được trong phiên giao dịch.
# Low: Giá thấp nhất mà tài sản rơi vào trong phiên giao dịch.
# Close: Giá đóng cửa của tài sản, tức là giá cuối cùng mà tài sản được giao dịch trong ngày. Đây là một chỉ số quan trọng được sử dụng rộng rãi trong phân tích kỹ thuật.
# Adj Close (Adjusted Close): Giá đóng cửa đã được điều chỉnh, phản ánh giá sau khi tính đến các yếu tố như cổ tức, phát hành thêm cổ phiếu, vv. Điều này giúp phân tích giá trị thực tế của tài sản qua thời gian.
# Volume: Khối lượng giao dịch, tức là tổng số lượng tài sản được mua và bán trong phiên giao dịch.
data.reset_index(drop=True, inplace=True)
```

```python
import plotly.graph_objects as go

figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], 
                                        high=data["High"],
                                        low=data["Low"], 
                                        close=data["Close"])])
figure.update_layout(title = "Phân Tích Giá Bitcoin theo USD trong vòng 5 năm", 
                     xaxis_rangeslider_visible=False)
figure.show()

figure_last_year = go.Figure(data=[go.Candlestick(x=data_last_year["Date"],
                                        open=data_last_year["Open"], 
                                        high=data_last_year["High"],
                                        low=data_last_year["Low"], 
                                        close=data_last_year["Close"])])
figure_last_year.update_layout(title = "Phân Tích Giá Bitcoin theo USD trong vòng 1 n gần đây", 
                     xaxis_rangeslider_visible=False)
figure_last_year.show()


figure_last_30days = go.Figure(data=[go.Candlestick(x=data_last_30days["Date"],
                                        open=data_last_30days["Open"], 
                                        high=data_last_30days["High"],
                                        low=data_last_30days["Low"], 
                                        close=data_last_30days["Close"])])
figure_last_30days.update_layout(title = "Phân Tích Giá Bitcoin theo USD trong vòng 1 tháng gần đây", 
                     xaxis_rangeslider_visible=False)
figure_last_30days.show()
```
