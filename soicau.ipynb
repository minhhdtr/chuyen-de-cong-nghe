```python
import json 

file_path = 'data.json'

# Đọc tệp JSON và chuyển đổi thành từ điển
with open(file_path, 'r') as file:
    json_data = json.load(file)



# In ra từ điển đã chuyển đổi
```

```python
import pandas as pd

for kq in json_data:
    kq['time'] = kq['time'].split(' ')[-1]
    kq['g2'] = ', '.join(kq['g2'])
    kq['g3'] = ', '.join(kq['g3'])
    kq['g4'] = ', '.join(kq['g4'])
    kq['g5'] = ', '.join(kq['g5'])
    kq['g6'] = ', '.join(kq['g6'])
    kq['g7'] = ', '.join(kq['g7'])
    
pd.DataFrame(json_data)
```

```python
import matplotlib.pyplot as plt

# Dữ liệu xổ số
giai_db = []
db_data = {}

for kq in json_data:
    giai_db.append(int(kq['db'][-2:])) 
    if int(kq['db']) in db_data:
        db_data[int(kq['db'][-2:])] += 1
    else:
        db_data[int(kq['db'][-2:])] = 1

ma_db = []
tansuat_db = []
for key, value in db_data.items():
    ma_db.append(key)
    tansuat_db.append(value)
    
# Tạo scatter plot
plt.scatter(ma_db, tansuat_db)

# Đặt tên trục x và y
plt.xlabel('mã')
plt.ylabel('tần suất')

# Đặt tiêu đề biểu đồ
plt.title('Biểu đồ tần suất giải đặc biệt từ ngày 01/11/2023 ~ 03/03/2024')

# Hiển thị biểu đồ
plt.show()
```
