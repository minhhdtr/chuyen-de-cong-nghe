```python
import json 

file_path = 'data.json'

# Đọc tệp JSON và chuyển đổi thành từ điển
with open(file_path, 'r') as file:
    json_data = json.load(file)



# In ra từ điển đã chuyển đổi
```

```python
import pandas as pd

for kq in json_data:
    kq['time'] = kq['time'].split(' ')[-1]
    kq['g2'] = ', '.join(kq['g2'])
    kq['g3'] = ', '.join(kq['g3'])
    kq['g4'] = ', '.join(kq['g4'])
    kq['g5'] = ', '.join(kq['g5'])
    kq['g6'] = ', '.join(kq['g6'])
    kq['g7'] = ', '.join(kq['g7'])
    
pd.DataFrame(json_data)
```
